```python
import pandas as pd
```

```python
# Загрузка данных из CSV файла с заданным форматом
data = pd.read_csv('movie.csv', sep=',')

data.head()
```

```python
# 1. Подсчет количества недостающих значений в поле 'duration'
missing_duration_count = data['duration'].isnull().sum()

print(f'Количество недостающих значений в поле "duration": {missing_duration_count}')
```

```python
# 2. Замена недостающих значений на медиану

# Вычисление медианы поля 'duration'
median_duration = data['duration'].median()
print(f"Median for 'duration': {median_duration}")

data['duration'].fillna(median_duration, inplace=True)
data.head()
```

```python
# 3. Среднее количество жанров у фильмов

# Разбиение строк с жанрами на списки
data['genres_list'] = data['genres'].str.split('|')

# Вычисление среднего количества жанров
average_genre_count = data['genres_list'].apply(len).mean()

print(f'Среднее количество жанров у фильмов: {average_genre_count:.2f}')
```

```python
# 4. Фильм/фильмы с максимальным количеством жанров.

# Разбиение строк с жанрами на списки
data['genres_list'] = data['genres'].str.split('|')

# Добавление нового столбца с количеством жанров
data['genre_count'] = data['genres_list'].apply(len)

# Нахождение максимального количества жанров
max_genre_count = data['genre_count'].max()

# Выбор фильмов с максимальным количеством жанров
movies_with_max_genres = data[data['genre_count'] == max_genre_count]

# Вывод информации о фильмах с максимальным количеством жанров
for index, movie in movies_with_max_genres.iterrows():
    print(f"Фильм: {movie['movie_title']}, Количество жанров: {max_genre_count}")
```

```python
# 5. Самый часто встречаемый жанр.

# Разбиение строк с жанрами на списки
data['genres_list'] = data['genres'].str.split('|')

# Создание плоского списка жанров
all_genres = [genre for genres_list in data['genres_list'] for genre in genres_list]

# Создание Series из списка жанров
genres_series = pd.Series(all_genres)

# Нахождение самого часто встречаемого жанра
most_common_genre = genres_series.value_counts().idxmax()
most_common_genre_count = genres_series.value_counts().max()

print(f"Самый часто встречаемый жанр: {most_common_genre} (встречается {most_common_genre_count} раз)")
```

```python
# 6. Сводная таблица о суммарном бюджете фильмов в разрезе стран за год, в 2014-2016 годах включительно.

# Фильтрация данных по годам 2014-2016
filtered_data = data[(data['title_year'] >= 2014) & (data['title_year'] <= 2016)]

# Построение сводной таблицы с суммарным бюджетом фильмов по странам и годам
pivot_table = pd.pivot_table(filtered_data, values='budget', index='country', columns='title_year', aggfunc='sum', fill_value=0)

pivot_table
```
